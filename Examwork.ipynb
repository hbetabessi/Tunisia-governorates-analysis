```python
#Create a new column 'IDR' (Regional Development Index) by aggregating the provided indices for each region.
```

```python
df['IDR'] = df[['Indice du savoir', 'Indice de la Richesse et Emploi', 'indice santé et population','Indice Justice et Equité']].mean(axis=1)
```

```python
df.head(5)
```

```python
#Calculate the IDR using a weighted average or a specific formula based on the provided indices.
```

```python
import numpy as np
```

```python
cols = ['Indice du savoir', 'Indice de la Richesse et Emploi',
        'indice santé et population', 'Indice Justice et Equité']
weights = np.array([0.3, 0.4, 0.2, 0.1])
```

```python
df['IDR'] = np.average(df[cols], weights=weights, axis=1)
```

```python
import matplotlib.pyplot as plt

indices = ['Indice du savoir', 'Indice de la Richesse et Emploi', 
           'indice santé et population', 'Indice Justice et Equité']

pairs = [(0,1), (0,2), (0,3), (1,2), (1,3), (2,3)]  

fig, axs = plt.subplots(2, 3, figsize=(18,10))
colors = ['green','blue','orange','purple','red','cyan']

for i, (x_idx, y_idx) in enumerate(pairs):
    row = i//3
    col = i%3
    axs[row,col].scatter(df[indices[x_idx]], df[indices[y_idx]], color=colors[i], s=50)
    axs[row,col].set_xlabel(indices[x_idx])
    axs[row,col].set_ylabel(indices[y_idx])
    axs[row,col].set_title(f"{indices[x_idx]} vs {indices[y_idx]} pour toutes les régions")
    axs[row,col].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
```

```python
#verify IDR exists
if 'IDR' not in df.columns:
    raise ValueError("La colonne 'IDR' n'existe pas dans df.")
```

```python
carte_tunisie =gpd.read_file('Tunisia_Governorates.geojson')
```

```python
print(carte_tunisie.columns)
```

```python
# Standardize names for the merge (all uppercase)
df['Gouvernorat_upper'] = df['Gouvernorat'].str.strip().str.upper()
carte_tunisie['NAME_EN_upper'] = carte_tunisie['NAME_EN'].str.strip().str.upper()
```

```python
# Check differences between names to ensure merge will work
print("Difference:", set(df['Gouvernorat_upper']) - set(carte_tunisie['NAME_EN_upper']))
```

```python
donnees_carte = carte_tunisie.merge(df, left_on='NAME_EN', right_on='Gouvernorat')
```

```python
print(donnees_carte[['NAME_EN','IDR']])
```

```python
# Perform the merge between the GeoJSON and the DataFrame
donnees_carte = carte_tunisie.merge(df, left_on='NAME_EN_upper', right_on='Gouvernorat_upper')
```

```python
# display card 
fig, ax = plt.subplots(1, 1, figsize=(12,12))
donnees_carte.plot(
    column='IDR',
    cmap='Reds',
    edgecolor='black',
    legend=True,
    ax=ax
)

ax.set_title("Indice de développement régional (IDR) par gouvernorat", fontsize=16)
ax.axis('off')
plt.show()
```

```python
# Plot the choropleth map
fig, ax = plt.subplots(1, 1, figsize=(12,12))
donnees_carte.plot(
    column='IDR',
    cmap='Reds',
    edgecolor='black',
    legend=True,
    ax=ax
)


for idx, row in donnees_carte.iterrows():
  
    centroid = row['geometry'].centroid
    ax.text(centroid.x, centroid.y, f"{row['IDR']:.2f}", 
            horizontalalignment='center', 
            verticalalignment='center', 
            fontsize=9, color='black', fontweight='bold')

ax.set_title("Indice de Développement Régional (IDR) par Gouvernorat", fontsize=16)

ax.axis('off') 

plt.show()
```

```python
#Add a legend and appropriate labels to enhance the map visualization.
```

```python
# Plot the choropleth map
fig, ax = plt.subplots(1, 1, figsize=(12,12))
choropleth = donnees_carte.plot(
    column='IDR',
    cmap='Reds',
    edgecolor='black',
    legend=True,
    ax=ax
)
# Add IDR labels at the centroid of each region
for idx, row in donnees_carte.iterrows():
    centroid = row['geometry'].centroid
    ax.text(centroid.x, centroid.y, f"{row['IDR']:.2f}", 
            horizontalalignment='center', 
            verticalalignment='center', 
            fontsize=9, color='black', fontweight='bold')

# Add a title
ax.set_title("Indice de Développement Régional (IDR) par Gouvernorat", fontsize=16)

# Enhance the legend
leg = choropleth.get_legend()
leg.set_title("IDR")  # Title for the colorbar

# Remove axes for clarity
ax.axis('off')

plt.show()
```
